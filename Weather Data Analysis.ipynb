# Analysis

# Pairplot for Multiple Variable Relationships

```python
# Create a pairplot to visualize relationships between multiple variables
sns.set(style="ticks")
sns.pairplot(df[['Temp_C', 'Rel Hum_%', 'Wind Speed_km/h', 'Press_kPa']], diag_kind="kde")
plt.show()
```

```python
#Analyze temperature trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Month", y="Temp_C")
plt.title("Temperature Trends")
plt.xlabel("Date/Time")
plt.ylabel("Temperature (Â°C)")
plt.xticks(rotation=45)
plt.show()
```

# Highest temperature

```python
# Find the date with the highest temperature
max_temp_date = df.loc[df['Temp_C'].idxmax()]['Date/Time']
max_temp = df['Temp_C'].max()
print(f"Highest Temperature ({max_temp}Â°C) occurred on {max_temp_date}")
```

# lowest temperature

```python
# Find the date with the lowest temperature
min_temp_date = df.loc[df['Temp_C'].idxmin()]['Date/Time']
min_temp = df['Temp_C'].min()
print(f"Lowest Temperature ({min_temp}Â°C) occurred on {min_temp_date}")
```

# Occurrences of different weather conditions by month

```python
# Count occurrences of different weather conditions by month
weather_condition_counts_by_month = df.groupby(['Month', 'Weather']).size().unstack().fillna(0)

# Plot a stacked bar chart of weather condition frequencies by month
plt.figure(figsize=(12, 6))
ax = weather_condition_counts_by_month.plot(kind='bar', stacked=True)
plt.title("Weather Condition Frequencies by Month")
plt.xlabel("Month")
plt.ylabel("Frequency")
plt.xticks(rotation=45)

# Adjust the legend's placement
plt.legend(title='Weather Condition', loc='upper left', bbox_to_anchor=(1, 1))

plt.show()
```

# Simple moving average for temperature

```python
# Calculate a simple moving average for temperature
window = 7  # Adjust the window size as needed
df['Temp_C_SMA'] = df['Temp_C'].rolling(window=window).mean()

# Plot temperature with its simple moving average
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date/Time', y='Temp_C', label='Temperature')
sns.lineplot(data=df, x='Date/Time', y='Temp_C_SMA', label=f'Temperature SMA ({window}-day)')
plt.title("Temperature and Simple Moving Average")
plt.xlabel("Date/Time")
plt.ylabel("Temperature (Â°C)")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

# Correlations between weather variables

```python
# Calculate correlations between weather variables
correlation_matrix = df.corr()

# Create a heatmap of correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title("Correlation Heatmap")
plt.show()
```

# Seasonal Trends in Relative Humidity

```python
# Calculate average relative humidity by month
monthly_avg_relative_humidity = df.groupby('Month')['Rel Hum_%'].mean()

# Create a bar chart to visualize seasonal trends in relative humidity
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_avg_relative_humidity.index, y=monthly_avg_relative_humidity.values)
plt.title("Seasonal Trends in Relative Humidity")
plt.xlabel("Month")
plt.ylabel("Average Relative Humidity (%)")
plt.show()
```

# Temperature vs. Dew Point Temperature

```python
# Create a scatter plot of temperature vs. dew point temperature
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Temp_C', y='Dew Point Temp_C')
plt.title("Temperature vs. Dew Point Temperature")
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Dew Point Temperature (Â°C)")
plt.show()
```

```python
# Plot wind speed trends over time using a line chart
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date/Time', y='Wind Speed_km/h')
plt.title("Wind Speed Trends Over Time")
plt.xlabel("Date/Time")
plt.ylabel("Wind Speed (km/h)")
plt.xticks(rotation=45)
plt.show()
```
