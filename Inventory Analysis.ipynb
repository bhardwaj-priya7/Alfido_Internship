# Distribution of purchase prices

```python
# Distribution of purchase prices
plt.figure(figsize=(10, 6))
sns.histplot(pur_p['SalesPrice'], bins=20, kde=True)
plt.title("Distribution of Purchase Prices")
plt.xlabel("Purchase Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# trend in purchase prices over time

plt.figure(figsize=(12, 6))
sns.lineplot(x='SalesDate', y='SalesPrice', data=pur_p)
plt.title("Trend in Purchase Prices Over Time")
plt.xlabel("Sales Date")
plt.ylabel("Purchase Price")
plt.xticks(rotation=45)
plt.show()
```
